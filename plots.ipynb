```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [6,7,8]
```

```python
fig0 = plt.figure(figsize = (5,3))  # an empty figure with no Axes, figsize is optional
```

```python
fig0 = plt.figure(figsize = (5,3))  # an empty figure with no Axes, figsize is optional
plt.plot(x,y)
```

```python
fig1, ax = plt.subplots(figsize = (5,7))  # a figure with a single Axes, figsize is optional
ax.set_title('single axes')
```

```python
fig1, ax = plt.subplots(figsize = (5,7))  # a figure with a single Axes, figsize is optional
ax.set_title('single axes')
ax.plot(x,y)
```

```python
fig2, axs = plt.subplots(2, 2, figsize = (10,8))  # a figure with a 2x2 grid of Axes, figsize is optional
axs[0,0].set_title('axs[0,0]')
axs[0,1].set_title('axs[0,1]')
axs[1,0].set_title('axs[1,0]')
axs[1,1].set_title('axs[1,1]')
```

```python
fig2, axs = plt.subplots(2, 2, figsize = (10,8))  # a figure with a 2x2 grid of Axes, figsize is optional
axs[0,0].set_title('axs[0,0]')
axs[0,0].plot(x,y)

axs[0,1].set_title('axs[0,1]')

axs[1,0].set_title('axs[1,0]')
axs[1,0].set_ylabel('ylabel')
axs[1,0].set_xlabel('xlabel')

axs[1,1].set_title('axs[1,1]')
axs[1,1].plot(x,y)
```
